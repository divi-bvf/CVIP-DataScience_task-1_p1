# Data Cleaning

```python
a=df.dropna(axis=1)
a
```

```python
for i in df.columns:
    print(i,end=',')
```

```python
col_keep= df[['iyear','imonth','iday','country','region','city','latitude','longitude','success','suicide','attacktype1','attacktype1_txt','targtype1_txt','gname','motive','weaptype1_txt','nkill','nwound','natlty1_txt']]
```

```python
df=col_keep.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country':'Country','region':'Region','city':'City','latitude':'Latitude','longitude':'Longitude','success':'Success','suicide':'Suicide','attacktype1':'AttackType','attacktype1_txt':'AttackType1_txt','targtype1_txt':'Target','gname':'Gang_Name','motive':'Motive','weaptype1_txt':'Weapon_Type','nkill':'Killed','nwound':'Wounded','natlty1_txt':'Country_Name'})
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#filling the killed and wounded null value if any with 0
df['Killed']=df['Killed'].fillna(0)
df['Wounded']=df['Wounded'].fillna(0)
```

```python
df.isnull().sum()
```

```python
#Countrywise distribution attack over the year
regg=pd.crosstab(df.Country_Name,df.Year)
reg2=pd.crosstab(df.Year,df.Region)
```

```python
reg2
```

```python
regg.plot(kind='area',figsize=(15,6))
```

```python
# Total killed bar and region plot

kx=reg2.transpose()
kx['Killed']=kx.sum(axis=1)
ks=kx['Killed'].sort_values(ascending=False)
```

```python
ks.plot(kind='bar',figsize=(15,6))
```

```python
#Country wise killed people barplot
regg['Killed']=regg.sum(axis=1)
j=regg['Killed'].sort_values(ascending=False)
j
```

```python
j.plot(kind='bar',figsize=(15,6))
```

```python
#terror attck type country and region wise
type_Atk=pd.crosstab(df.AttackType1_txt,df.Year)
```

```python
type_Atk
```

```python
cty_type=type_Atk.transpose()
cty_type['Country']=cty_type.sum(axis=1)
k=cty_type.sort_values(by='Country',ascending=False)
k
```

```python
k.plot(kind='bar',figsize=(15,6))
```

```python
reg_type=type_Atk.transpose()
reg_type['Region']=reg_type.sum(axis=1)
k2=reg_type.sort_values(by='Region',ascending=False)
k2
```

```python
k2.plot(kind='bar',figsize=(15,6))
```
